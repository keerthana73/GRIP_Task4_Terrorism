# EDA

```python
sns.heatmap(df.corr(),cmap='coolwarm')
plt.show()
```

```python
df['iyear'].value_counts().plot(kind='pie',figsize=(7,10)).invert_yaxis()
plt.title('Year with Max Attacks')
```

```python
df['imonth'].value_counts().plot(kind='pie').invert_yaxis()
```
